```python
# Imports
import gmaps
import numpy as np
import pandas as pd
import requests
import time
from apikey import api_key
from pprint import pprint
import matplotlib as plt
```

# Loblaws

```python
# Search Parameters for Loblaws Center + Request Page 1 
target_coordinates = "43.739541, -79.356733"
target_search = ("Loblaws")
target_radius = 13000
target_type = ("Supermarket")

params = {
    "location": target_coordinates,
    "keyword": target_search,
    "radius": target_radius,
    "type": target_type,
    "key": api_key
    }

# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
response=requests.get(base_url, params=params)
```

```python
result_list=[]
flag = True
count=0
while flag:
    count+=1
    print("Working on Page ",count)


    loblaws_data = response.json()
    lat_loblaws = []
    lng_loblaws = []
    results = loblaws_data['results']
    loblaws_geo = pd.DataFrame(results)['geometry']
    loblaws_title = pd.DataFrame(results)['name']
    for x in loblaws_geo:
        loblaws_location = pd.DataFrame(x)['location']
        lat_loblaws.append(loblaws_location['lat'])
        lng_loblaws.append(loblaws_location['lng'])
    

    loblaws_lat_pd = pd.DataFrame(lat_loblaws)
    loblaws_lng_pd = pd.DataFrame(lng_loblaws)
    loblaws_title_pd = pd.DataFrame(loblaws_title)
    loblaws_title_pd.join(loblaws_lat_pd)
    loblaws_title_pd['C'] = loblaws_title_pd.reset_index().index
    loblaws_lat_pd['C'] = loblaws_lat_pd.reset_index().index
    loblaws_lng_pd['C'] = loblaws_lng_pd.reset_index().index
    sample2=pd.merge(loblaws_title_pd,loblaws_lat_pd,on='C')
    sample1=pd.merge(sample2,loblaws_lng_pd,on='C')
    result_list.append(sample1)

    if('next_page_token' in loblaws_data):
        paramaters = {
        "key": api_key,
        "pagetoken":loblaws_data['next_page_token']
        }
        time.sleep(2)

        response=requests.get(base_url, paramaters)
    else:
        flag=False

```

```python
result_df = pd.concat(result_list, ignore_index=True)[['name','0_x','0_y']]

loblaws_locations_pd = pd.DataFrame(result_df)
loblaws_locations_pd.iloc[0:19]
```
