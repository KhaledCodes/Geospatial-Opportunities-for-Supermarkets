```python
# Generate the Forward Sortation Area (FSA) we are interested (M-series) so we can query them in QGIS and 
# create a GeoJSON file to input into Folium to create borders for each FSA
fsa_list = []
for i in range(1,10):
    fsa = "M" + str(i)
    for a in range(65,91):
        fsaf = fsa + str(chr(a))
        fsa_list.append(fsaf)
```

```python
import folium # map rendering library
import json 
import numpy as np
import pandas as pd
```

```python
df_pop = pd.read_csv("T1201EN.CSV")
# we can immediately drop the columns we're not interested in 
df_pop=df_pop[['Geographic code','Province or territory', 'Population, 2016']]
df_pop=df_pop[(df_pop['Province or territory']== "Ontario")]
df_pop.head()
```

```python
map_toronto = folium.Map(location=[43.653963, -79.387207], zoom_start=10)
ontario_geo = "Toronto FSAs.geojson"
map_toronto.choropleth(geo_data=ontario_geo,
    data = df_pop,
    columns=['Geographic code','Population, 2016'],
    key_on='feature.properties.CFSAUID',
    fill_color='YlOrRd',
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Population by FSA')   
    
map_toronto
```
